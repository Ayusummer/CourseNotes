## 实验 2-理论题作答

总开销 E (千元)   
总月数 m  
自己开发策略总开销: E1 = 30 + 3.5m  
购买软件公司产品策略总开销: E2 = 18 + 4.2m

```python
from matplotlib import pyplot as plt
m = list(range(1, 20, 1))
E1 = [30 + 3.5 * i for i in m]
E2 = [18 + 4.2 * i for i in m]
x1 = 12 / 0.7
y1 = 30 + 3.5 * x1
plt.figure(figsize=(10, 5))
plt.xticks(m)
plt.plot(m, E1)
plt.plot(m, E2)
plt.scatter(x1, y1)
plt.axvline(x1, color='magenta', linestyle='--', linewidth=2)
plt.show()
```

## 实验 8 - 计算 ROI


```python
x1 = [-100000, 10000, 10000, 10000, 20000, 100000]
x2 = [-1000000, 200000, 200000, 200000, 200000, 300000]
x3 = [-100000, 30000, 30000, 30000, 30000, 30000]
x4 = [-120000, 30000, 30000, 30000, 30000, 75000]
def f(x):
    return x[0] + x[1] * 0.9091 + x[2] * 0.8264 + x[3] * 0.7513 + x[4] * 0.6830 + x[5] * 0.6209
print("NPV1 = {0}, NPV2 = {1}, NPV3 = {2}, NPV4 = {3}".format(f(x1), f(x2), f(x3), f(x4)))
```
